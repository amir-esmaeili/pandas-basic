```python
# courses which are free and have the most number of subscribers
subs_free = df.loc[(df.is_paid == False)]
subs_free.sort_values('num_subscribers', ascending=False).head()
```

```python
# poplular courses
df['popularity'] = df.num_subscribers + df.num_reviews
df.sort_values('popularity', ascending=False).head()
```

```python
df.plot.scatter(x='popularity', y='price', figsize=(8,6))
```

```python
def get_free_courses(subject_name:str) -> pd.DataFrame:
    df_free = df[(df.is_paid == False) & (df.subject == subject_name)]
    df_free.drop(['is_paid', 'price', 'published_timestamp'], axis=1, inplace=True)
    return df_free.sort_values('num_subscribers', ascending=False).head()
```

```python
get_free_courses('Business Finance')
```
