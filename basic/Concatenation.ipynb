```python
import pandas as pd
```

```python
Team1 = {
    'Player': ['Jack', 'Mat', 'Steve', 'Tyler', 'Eric'],
    'Pos': ['CF', 'SS', 'MC', 'DFB', 'GK'],
    'Score': [75, 80, 73, 64, 70]
}
Team2 = {
    'Player': ['Alex', 'Daniel', 'Fred', 'Tom', 'Mark'],
    'Pos': ['CF', 'MCD', 'MC', 'DFB', 'GK'],
    'Score': [68, 86, 64, 59, 70]
}
team1_df = pd.DataFrame(Team1)
team2_df = pd.DataFrame(Team2)
```

```python
team1_df
```

```python
team2_df
```

```python
# concate 
pd.concat([team1_df, team2_df], ignore_index=True) # ignore index for indexing continuously
```

```python
pd.concat([team1_df, team2_df], keys=['Team1', 'Team2'])
```

```python
# horizontal concate
pd.concat([team1_df, team2_df], axis=1, keys=['Team1', 'Team2'])
```

```python
# basic approach
team1_df.append([team2_df])
```
