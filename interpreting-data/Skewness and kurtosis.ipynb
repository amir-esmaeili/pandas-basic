```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
dataset = pd.read_csv('./datasets/500_Person_Gender_Height_Weight_Index.csv')
dataset.head()
```

```python
dataset.drop('Index', inplace=True, axis=1)
dataset.shape
```

```python
dataset.Weight.plot(kind="kde", title = "Weight", figsize=(12, 8))
plt.show()
```

```python
dataset.Height.plot(kind="kde", title="Height", figsize=(12, 8))
plt.show()
```

```python
dataset.skew()
```

```python
dataset.kurtosis()
```
