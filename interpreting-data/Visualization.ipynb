```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('ggplot')
```

```python
dataset = pd.read_csv('./datasets/weight-height.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(dataset.Gender)
plt.show()
```

```python
height = dataset.Height
height.head()
```

```python
plt.figure(figsize=(12,8))
height.plot(kind = "hist", title = "Height Histogram")
plt.show()
```

```python
plt.figure(figsize=(12,8))
height.plot(kind = "box", title = "Height Box-plot (inch)")
plt.show()
```

```python
# probability distribution
height.plot(kind = "kde", title = "Height KDE", figsize=(12,8))
plt.show()
```

```python
weight = dataset.Weight
weight.head()
```

```python
weight.shape
```

```python
# bi-mode distribution
weight.plot(kind = "kde", title = "Weight KDE", figsize=(12, 8))
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = "Height", y = "Weight", data = dataset)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = "Height", y = "Weight", hue = "Gender", data = dataset)
plt.show()
```

```python
gender_groups = dataset.groupby("Gender", as_index=False)
gender_groups.head()
```

```python
gender_groups.describe().T
```

```python
sns.FacetGrid(dataset, hue="Gender", height=5).map(sns.distplot, "Height").add_legend()
plt.show()
```

```python
sns.FacetGrid(dataset, hue="Gender", height=5).map(sns.distplot, "Weight").add_legend()
plt.show()
```
